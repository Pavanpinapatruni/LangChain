## Cosine Similarity with OpenAI Embeddings

```python
sentences = [
    "The cat is on the roof",
    "A dog is barking in the yard",
    "I love programming in Python",
    "RAG systems combine retrieval and generation for powerful applications",
    "Langchain is a great framework for building RAG applications"
]
```

```python
def cosine_similarity(vec1, vec2):
    """
    Cosine similarity measures the angle between two vectors.
    - Result close to 1: Very similar
    - Result close to 0: Not related
    - Result close to -1: Opposite meanings
    """
    dot_product = np.dot(vec1, vec2)
    norm_a = np.linalg.norm(vec1)
    norm_b = np.linalg.norm(vec2)
    return dot_product/(norm_a * norm_b)
```

```python
from langchain_openai import OpenAIEmbeddings

# 1. Setup credentials
GITHUB_TOKEN = os.environ.get("GITHUB_API_KEY")
ENDPOINT = "https://models.inference.ai.azure.com"

# 2. Use OpenAIEmbeddings with GitHub Models endpoint (OpenAI-compatible API)
embeddings = OpenAIEmbeddings(
    model="text-embedding-3-small",
    openai_api_key=GITHUB_TOKEN,
    openai_api_base=ENDPOINT,
)

embeddings
```

```python
sentence_embeddings = embeddings.embed_documents(sentences)
```

```python
import numpy as np
## Calculate cosine similarity between all the pairs of sentences
for i in range(len(sentences)):
    for j in range(i + 1, len(sentences)):
        sim = cosine_similarity(sentence_embeddings[i], sentence_embeddings[j])
        print(f"Sentences: '{sentences[i]}' vs '{sentences[j]}'")
        print(f"Similarity between '{sentences[i]}' and '{sentences[j]}': {sim:.4f}\n")
```

```python
### Example- Semantic Search- Retrieve the similar sentence
# Test semantic search
documents = [
    "LangChain is a framework for developing applications powered by language models",
    "Python is a high-level programming language",
    "Machine learning is a subset of artificial intelligence",
    "Embeddings convert text into numerical vectors",
    "The weather today is sunny and warm"
]
query="What is Langchain?"
```

```python
def semantic_search(query, documents, embeddings, top_k=3):

    # Embed the query and documents
    query_embedding = embeddings.embed_query(query)
    document_embeddings = embeddings.embed_documents(documents)
    
    similarties=[]

    for i,doc_emb in enumerate(document_embeddings):
        similarity=cosine_similarity(query_embedding,doc_emb)
        similarties.append((similarity,documents[i]))

    ## Sort by similarity
    similarties.sort(reverse=True)
    return similarties[:top_k]
```

```python
results = semantic_search(query, documents, embeddings)
print("Most similar documents to the query:")
print(results)
```
