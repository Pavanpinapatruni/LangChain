### Word Document Processing

```python
from langchain_community.document_loaders import Docx2txtLoader, UnstructuredWordDocumentLoader
```

### Method1 : Using Docx2txtLoader

```python
try:
    docx_loader = Docx2txtLoader("./data/word_files/proposal.docx")
    docs = docx_loader.load()
    print(f"Loaded {len(docs)} documents from Word file.")
    print(f"Content Preview:\n{docs[0].page_content[:200]}...")
    print(f"metadata: {docs[0].metadata}")
except Exception as e:
    print(f"Error loading Word document: {e}")
```

### Method2 : Using UnstructuredWordDocumentLoader

```python
try:
    unstructured_loader = UnstructuredWordDocumentLoader("./data/word_files/proposal.docx", mode="elements")
    unstructured_docs = unstructured_loader.load()
    print(f"Loaded {len(unstructured_docs)} documents from proposal.docx using Docx2txtLoader.")

    for i, doc in enumerate(unstructured_docs[:2]):
        print(f"metadata: {doc.metadata.get('category', 'unknown')}")
        print(f"Element {i+1} Content Preview:\n{doc.page_content[:100]}...")
except Exception as e:
    print(f"Error loading proposal.docx with Docx2txtLoader: {e}")
```

**Note** : If you observe because of the parameter mode = "elements" already data is splitted into chunks.  
If you want, you can again split into chunks by regular methods
